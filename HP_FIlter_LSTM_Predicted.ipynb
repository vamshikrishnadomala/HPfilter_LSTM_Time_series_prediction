```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_excel(r"51000_Decom.xlsx")
```

```python
dataset
```

```python
df=dataset[39480:]
```

```python
fig = plt.figure(figsize=(18, 6), dpi=100)
plt.plot(df['Date'],df['WVHT'], label='Measured', color='red')
plt.plot(df['Date'], df['Predicted'], label='Test_Predicted', color='green')
plt.title("LSTM Wave height prediction")
plt.legend(loc='upper right')
plt.show()
```
